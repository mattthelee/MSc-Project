```python
import dataEnv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pdb
```

```python
mission = 'MineRLNavigateDense-v0'
env = dataEnv.DataEnv(mission,limit = -1, filters=False)
steps = []
for metaData in env.meta:
    if metaData['total_reward'] >= 164:
        steps.append(metaData['duration_steps'])
mission = 'MineRLNavigate-v0'
env = dataEnv.DataEnv(mission,limit = -1, filters=False)
for metaData in env.meta:
    if metaData['total_reward'] >= 100:
        steps.append(metaData['duration_steps'])
stepsDf = pd.DataFrame(steps)
del steps
```

```python
sns.distplot(stepsDf[0],50)
plt.ylabel('Probability')
plt.xlabel('Steps')
plt.xlim(left=0)
print(stepsDf.mean())
print(stepsDf.mode())
print(stepsDf.max())
```

```python
mission = 'MineRLNavigateDense-v0'
env = dataEnv.DataEnv(mission,limit = -1, filters=False)
obsActionList = env.obsActionList
del env
mission = 'MineRLNavigate-v0'
env = dataEnv.DataEnv(mission,limit = -1, filters=False)
obsActionList.extend(env.obsActionList)
del env

# intialise countDict
countDict = {}
for key in obsActionList[0][1].keys():
    if key == 'camera':
        continue
    countDict[key] = 0

countDict['lookLeft'] = 0
countDict['lookRight'] = 0
countDict['lookUp'] = 0
countDict['lookDown'] = 0
for obs, action in obsActionList:
    for key in action.keys():
        if key == 'camera':
            if action[key][0] > 3:
                countDict['lookRight'] += 1./len(obsActionList)
            if action[key][0] < -3:
                countDict['lookLeft'] += 1./len(obsActionList)
            if action[key][1] > 3:
                countDict['lookUp'] += 1./len(obsActionList)
            if action[key][1] < -3:
                countDict['lookDown'] += 1./len(obsActionList)
            continue
        if action[key] > 0:
            countDict[key] += 1./len(obsActionList)

    
```

```python
plt.bar(range(len(countDict)), list(countDict.values()))
plt.xticks(range(len(countDict)), list(countDict.keys()), rotation=90)
plt.xlabel('Sub-action')
plt.ylabel('Fraction of actions containing subaction')
```
